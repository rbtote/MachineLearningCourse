```python
translator = Translator()
nltk.download('wordnet')
```

```python
#create a list of stop words (SET)
stop_words = set(stopwords.words("spanish"))

stop_words.add('clase')
stop_words.add('clases')
stop_words.add('tema')
stop_words.add('temas')
stop_words.add('materia')
stop_words.add('profesor')
stop_words.add('profesora')
stop_words.add('maestra')
stop_words.add('maestro')

#show how many words are in the list of stop words
print(len(stop_words))
```

```python
#construct a new list to store the cleaned text
clean_desc = []
for w in range(len(df_analysis.Comentarios)):
    desc =df_analysis.iloc[w]['Comentarios'].lower()
    
    #remove punctuation
    desc = re.sub('[^a-zA-ZñÑáÁéÉíÍóÓúÚ]', ' ', desc)
    
    #remove tags
    desc = re.sub("&lt;/?.*?&gt;"," &lt;&gt; ",desc)
    
    #remove special characters and digits
    desc = re.sub("(\\d|\\W)+"," ",desc)
    
    split_text = desc.split()
    
    #Lemmatisation
    lem = WordNetLemmatizer()
#   split_text = [lem.lemmatize(word) for word in split_text if not word in stop_words and len(word) >2] 
    split_text = [word for word in split_text if not word in stop_words and len(word) >2] 
    split_text = " ".join(split_text)
    clean_desc.append(split_text)
```

```python
#TF-IDF vectorizer
tfv = TfidfVectorizer(stop_words = stop_words, ngram_range = (1,1))
#transform
vec_text = tfv.fit_transform(clean_desc)
#returns a list of words.
words = tfv.get_feature_names()
```

```python
veclen = vec_text.shape[0]
veclen = math.floor(veclen*.8)
if(veclen > 50):
    veclen = 30
```

# Check best number of clusters

```python
sse = []

for i in range(1, veclen):
    kmeans = KMeans(n_clusters = i, random_state = 0)
    kmeans.fit(vec_text)
    sse.append(kmeans.inertia_)
    
#Plotting elbow method 
plt.plot(range(1, veclen), sse)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
```

```python
#setup kmeans clustering

clustersWords = []
#                    CHANGE NUMBER OF CLUSTERS
#                             v
#                           #### 
kmeans = KMeans(n_clusters = 16, n_init = 17, tol = 0.01, max_iter = 200)
#fit the data 
kmeans.fit(vec_text)
#this loop transforms the numbers back into words
common_words = kmeans.cluster_centers_.argsort()[:,-1:-11:-1]
for num, centroid in enumerate(common_words):
    print(str(num) + ' : ' + ', '.join(words[word] for word in centroid))
    clustersWords.append('' + ', '.join(words[word] for word in centroid))
```

```python
#add the cluster label to the data frame
df_analysis['cluster'] = kmeans.labels_
clusters = df_analysis.groupby(['cluster', 'DocumentSentiment']).size()
fig, ax1 = plt.subplots(figsize = (26, 15))
sns.heatmap(clusters.unstack(level = 'DocumentSentiment'), ax = ax1, cmap = 'Reds')
ax1.set_xlabel('DocumentSentiment').set_size(18)
ax1.set_ylabel('cluster').set_size(18)
clusters = df_analysis.groupby(['cluster', 'above_avg']).size()
fig2, ax2 = plt.subplots(figsize = (30, 15))
sns.heatmap(clusters.unstack(level = 'above_avg'), ax = ax2, cmap="Reds")
ax2.set_xlabel('Above Average DocumentSentiment').set_size(18)
ax2.set_ylabel('Cluster').set_size(18)
```

```python
#create dataframe of docSentiment not above average
not_above = df_analysis.loc[df_analysis['above_avg'] == 0]
not_above.describe()

#create data frame of reviews above average
above_avg = df_analysis.loc[df_analysis['above_avg'] == 1]
above_avg.describe()

#plot the counts
plt.figure(figsize=(14,4))
sns.countplot(x='cluster', data=not_above).set_title("NEGATIVE DocumentSentiment")
plt.show()
plt.figure(figsize=(14,4))
sns.countplot(x='cluster', data=above_avg).set_title("POSITIVE DocumentSentiment")
plt.show()
```

```python
Neg_clusters = set(not_above['cluster'].tolist())

Pos_clusters = set(above_avg['cluster'].tolist())
```

```python
GoodTeachers = list(Pos_clusters.difference(Neg_clusters))

BadTeachers = list(Neg_clusters.difference(Pos_clusters)) 
```

# Good teachers are/have/do

```python
for cluster in GoodTeachers:
    print(clustersWords[cluster])
```
