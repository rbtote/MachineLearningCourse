```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from sklearn.metrics import accuracy_score 

plt.style.use('ggplot')
%matplotlib inline
```

```python
iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target
```

```python
C = 1.0  # SVM regularization parameter
models = (svm.LinearSVC(C=C, max_iter=10000),
          svm.SVC(kernel='linear', C=C),          
          svm.SVC(kernel='rbf', gamma=0.7, C=C),
          svm.SVC(kernel='poly', degree=3, gamma='auto', C=C))
```

```python
models = (model.fit(X, y) for model in models)
```

```python
def make_meshgrid(x, y, h=.02):
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    return xx, yy
```

```python
def plot_contours(ax, model, xx, yy, **params):
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out
```

```python
# title for the plots
titles = ('LinearSVC',
          'SVC linear kernel',          
          'SVC RBF kernel',
          'SVC polynomial (degree 3) kernel')

# Set-up 2x2 grid for plotting.
fig, sub = plt.subplots(2, 2, figsize=(15,15))
plt.subplots_adjust(wspace=0.8, hspace=0.8)

X0, X1 = X[:, 0], X[:, 1]
xx, yy = make_meshgrid(X0, X1)

for model, title, ax in zip(models, titles, sub.flatten()):
    plot_contours(ax, model, xx, yy, cmap=plt.cm.coolwarm)
    ax.scatter(X0, X1, c=y, cmap=plt.cm.coolwarm)
    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    ax.set_xlabel('Sepal length')
    ax.set_ylabel('Sepal width')
    ax.set_title(title)

plt.show()
```

```python
C = 1.0  # SVM regularization parameter
models = (svm.LinearSVC(C=C, max_iter=10000),
          svm.SVC(kernel='linear', C=C),          
          svm.SVC(kernel='rbf', gamma=0.7, C=C),
          svm.SVC(kernel='poly', degree=3, gamma='auto', C=C))
```

```python
for model, title in zip(models, titles):
    model.fit(X, y)
    classifier_predictions = model.predict(X)
    print(title,' : ',accuracy_score(y, classifier_predictions)*100)
```

As we can see, in this dataset, the model with the best accuracy is the RBF kernel one.
But looking closely to the plot, we can see that it is maybe not the best choice, as we can see some red areas in the left-most side, and we don't think that a point in the top left corner would be in that category. So it is probably having some overfitting.

We also can see that maybe for the blue category, a linear SVM would be enough, but for the gray and red ones we don't see a clear division, so for this time I would choice the polynomial SVC as the best fit for this dataset.
