# Clean

```python
#convertir los datos categoricos a valores numericos
def convert_values(label):
    column = pref[label]
    values = list()
    key = list()
    column_labels =dict(column.value_counts())
    for i,value in enumerate(column_labels):
        values.append(value)
        key.append(i)
    column = column.replace(values, key)
    return column
```

```python
pref['Alcohol'] = convert_values('Alcohol')
pref['Punctuality'] = convert_values('Punctuality')
pref['Lying'] = convert_values('Lying')
pref['Internet usage'] = convert_values('Internet usage')
pref['Gender'] = convert_values('Gender')
pref['Left - right handed'] = convert_values('Left - right handed')
pref['Education'] = convert_values('Education')
pref['Only child'] = convert_values('Only child')
pref['Village - town'] = convert_values('Village - town')
pref['House - block of flats'] = convert_values('House - block of flats')
pref['Smoking'] = convert_values('Smoking')
```

## Outliers

```python
def find_outliers_tukey(x):
    q1 = np.percentile(x, 25)
    q3 = np.percentile(x, 75)
    iqr = q3-q1 
    floor = q1 - 1.5*iqr
    ceiling = q3 + 1.5*iqr
    outlier_indices = list(x.index[(x < floor)|(x > ceiling)])
    outlier_values = list(x[outlier_indices])

    return outlier_indices, outlier_values
```

```python
tukey_indices, tukey_values = find_outliers_tukey(X['Age'])
print(np.sort(tukey_values))
```
