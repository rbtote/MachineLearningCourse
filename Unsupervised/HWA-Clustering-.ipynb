```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
from sklearn import metrics



#importing the Iris dataset with pandas
iris = load_iris()
X = iris['data']
y = iris['target']
```

```python
sse = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, random_state = 0)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)
    
#Plotting elbow method 
plt.plot(range(1, 11), sse)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
```

```python
silohuette = np.zeros(9)

for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state = 0).fit(X)
    results = kmeans.predict(X, sample_weight=None)
    silohuette[i-2] = metrics.silhouette_score(X, results)
    
#Plotting silohuette coefficient
plt.plot(range(2, 11), silohuette)
plt.title('Silohouette Coefficient')
plt.xlabel('Number of clusters')
plt.ylabel('Silohouette coefficient')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, random_state = 0)
y_kmeans = kmeans.fit_predict(X)
```

```python
#this will tell us to which cluster does the data observations belong.
new_labels = kmeans.labels_
# Plot the identified clusters and compare with the answers
fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].scatter(X[:, 0], X[:, 1], c=y, cmap='gist_rainbow', edgecolor='k', s=150)
axes[1].scatter(X[:, 0], X[:, 1], c=new_labels, cmap='jet', edgecolor='k', s=150)
axes[0].set_xlabel('Sepal length')
axes[0].set_ylabel('Sepal width')

axes[0].set_title('Actual')
axes[1].set_title('KMeans Predicted')
```

```python
dbscan = DBSCAN(eps=0.5, metric='euclidean', min_samples=5)
dbscan.fit(X)
```

We are using an euclidean distance of 0.5 and 5 samples in order to create a cluster, that parameters were estimated after some tests in the dataset.

```python
pca = PCA(n_components=2).fit(X)
pca_2d = pca.transform(iris.data)
for i in range(0, pca_2d.shape[0]):
    if dbscan.labels_[i] == 0:
        c1 = plt.scatter(pca_2d[i,0],pca_2d[i,1],c='r', marker='o')
    elif dbscan.labels_[i] == 1:
        c2 = plt.scatter(pca_2d[i,0],pca_2d[i,1],c='g', marker='o')
    elif dbscan.labels_[i] == -1:
        c3 = plt.scatter(pca_2d[i,0],pca_2d[i,1],c='b', marker='*')
plt.legend([c1, c2, c3], ['Cluster 1', 'Cluster 2', 'Noise'])
plt.title('DBSCAN Predicted')
plt.show()
```

In the case of the IRIS dataset, I think that it is more accurate to use KMeans, where we can specify the number of classes that we want (in this case because we already know). And in DBSCAN, we are getting just 2 clusters, because two of the real classes have not much difference.
