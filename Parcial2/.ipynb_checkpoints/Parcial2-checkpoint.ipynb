```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
%matplotlib inline

from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
import plotly.express as px
```

```python
df=pd.read_csv('publication_strategies.csv')
```

```python
df.head()
```

# Data exploration

# PCA

```python
df.shape
```

```python
# Se divide la matriz del dataset en dos partes
X = df[['affiliations','intl_afil','authors','docs','book','chapter','conference','journal','q1q2','q3q4']].values 
y = df[['cites_3y']].values
```

```python
#Transformación a una distribución normal
X_std = StandardScaler().fit_transform(X)
```

```python
#Matriz de covarianza
print('NumPy covariance matrix: \n%s' %np.cov(X_std.T))
```

```python
#Calcular autovalores y autovectores
cov_mat = np.cov(X_std.T)
eig_vals, eig_vecs = np.linalg.eig(cov_mat)
print('Eigenvectors \n%s' %eig_vecs)
print('\nEigenvalues \n%s' %eig_vals)
```

```python
# Hacer parejas (autovector, autovalor), y ordenarlas en orden descendiente
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]
eig_pairs.sort(key=lambda x: x[0], reverse=True)
print('Autovalores en orden descendiente:')
for i in eig_pairs:
    print(i[0])
```

```python
# A partir de los autovalores, calcular la varianza explicada
tot = sum(eig_vals)
var_exp = [(i / tot)*100 for i in sorted(eig_vals, reverse=True)]
cum_var_exp = np.cumsum(var_exp)

# Representamos en un diagrama de barras la varianza explicada por cada autovalor, y la acumulada
with plt.style.context('seaborn-pastel'):
    plt.figure(figsize=(6, 4))

    plt.bar(range(10), var_exp, alpha=0.5, align='center',
            label='Varianza individual explicada', color='g')
    plt.step(range(10), cum_var_exp, where='mid', linestyle='--', label='Varianza explicada acumulada')
    plt.ylabel('Ratio de Varianza Explicada')
    plt.xlabel('Componentes Principales')
    plt.legend(loc='best')
    plt.tight_layout()
```

```python
#Generamos la matríz a partir de los pares autovalor-autovector
matrix_w = np.hstack((eig_pairs[0][1].reshape(10,1),
                      eig_pairs[1][1].reshape(10,1),
                      eig_pairs[2][1].reshape(10,1),
                      eig_pairs[3][1].reshape(10,1)))

matrix_w[:,1] =-matrix_w[:,1]
matrix_w[:,2] =-matrix_w[:,2]
matrix_w[:,3] =-matrix_w[:,3]

print('Matriz W:\n', matrix_w)
Y = X_std.dot(matrix_w)
```

```python
PCdf = {
        'PC1': Y[:,0],
        'PC2': Y[:,1],
        'PC3': Y[:,2],
        'PC4': Y[:,3],
        'Cites': df['cites_3y']
        }
PCdf = pd.DataFrame(data=PCdf)
```

# K-means

```python
sse = []

for i in range(1, 9):
    kmeans = KMeans(n_clusters = i, random_state = 0)
    kmeans.fit(Y)
    sse.append(kmeans.inertia_)
    
#Plotting elbow method 
plt.plot(range(1, 9), sse)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
```

```python
silohuette = np.zeros(14)

for i in range(2,16):
    kmeans = KMeans(n_clusters=i, random_state = 0).fit(Y)
    results = kmeans.predict(Y, sample_weight=None)
    silohuette[i-2] = metrics.silhouette_score(Y, results)
    
#Plotting silohuette coefficient
plt.plot(range(2,16), silohuette)
plt.title('Silohouette Coefficient')
plt.xlabel('Number of clusters')
plt.ylabel('Silohouette coefficient')
plt.show()
```

```python
kmeans = KMeans(n_clusters=8,random_state=0).fit(Y)

results = kmeans.predict(Y, sample_weight=None)

PCdf["clusterKmeans8"] = results
print(results)

#metrics.silhouette_score(X, results)
```

# DBSCAN

```python
#from scipy.spatial.distance import braycurtis

dbscan = DBSCAN(eps=.68, metric='euclidean', min_samples=6)
dbscan.fit(Y)
dbscan_labels = dbscan.labels_
labels, counts = np.unique(dbscan_labels[dbscan_labels>=0], return_counts=True)
labelOut, outliers = np.unique(dbscan_labels[dbscan_labels==-1], return_counts=True)
PCdf["clusterDBSCAN"] = dbscan_labels
print(labels,counts)
print(labelOut,outliers)
print(outliers*100/(sum(counts)+outliers))
```

```python
fig = px.scatter(PCdf, x='PC1', y='PC2', color='clusterDBSCAN')
fig.show()
```

```python
PCdf.groupby('clusterDBSCAN').mean()
```
